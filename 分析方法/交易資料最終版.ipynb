```python
import pandas as pd
```

```python
df = pd.read_csv('transaction_data_version2.csv')
```

```python
df
```

```python
def time_trans(i):
    if len(str(i)) ==4:
        return str(i)[0:2]
    elif len(str(i)) ==3:
        return(str(i)[0])
    elif len(str(i)) ==2:
        return(str(0))
    elif len(str(i)) ==1:
        return(str(i))
```

```python
df['hour'] = df.TRANS_TIME.map(time_trans)
```

```python
df.head()
```

```python
import datetime as dt
```

```python
df['hour_time'] = pd.to_datetime(df['hour'],format= '%H' ).dt.time
```

```python
df2 = df[['household_key','MEMBER','BASKET_ID','PRODUCT_ID','QUANTITY','SALES_VALUE','WEEK_NO','DAY','DATE','hour_time']]
```
