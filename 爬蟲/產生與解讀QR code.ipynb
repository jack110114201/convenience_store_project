### 解讀qr code

### linebot 這邊應該就是消費者傳送商品qrcode圖片，我們再解讀，並讀取商品資料庫資料

```python
import cv2
# read the QRCODE image
img1 = cv2.imread("./qrcode_folder/product_25671.png") 
QRCodeDetector = cv2.QRCodeDetector()
retval, points, straight_qrcode = QRCodeDetector.detectAndDecode(img1)
print(retval) #Result in a string.
print('*'*100)
print(points) #Array of vertices of the found QR code quadrangle. Will be empty if not found.
print('*'*100)
print(straight_qrcode) #An image containing rectified and binarized QR code.
```
