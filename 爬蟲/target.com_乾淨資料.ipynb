### 測groupby

```python
df1 = pd.DataFrame({"name":["A","A","B","B","A"],"price":[np.nan,np.nan,np.nan,np.nan,np.nan]})
df2 = pd.DataFrame({"name":["A","A","B","B"],"price":[10,20,30,40]})
```

```python
df1_group = df1.groupby("name")
```

### 單次

```python
g1 = df1_group.get_group("A").reset_index(drop=True)
g2 = df2_group.get_group("A").reset_index(drop=True)
for j in range(df1_group.get_group("A").value_counts(subset="name")[0].item()):
    try:
        g1.iloc[j,1] = g2.iloc[j,1]
    except:
        g1.reset_index(drop=True).iloc[j,1] = np.nan

g1.to_csv("./temp_file/a_target.csv", index=False)
```

### 多次

```python
name_list = ["A","B"]
```

```python
for i in name_list:
    g1 = df1_group.get_group(i).reset_index(drop=True)
    g2 = df2_group.get_group(i).reset_index(drop=True)
    for j in range(df1_group.get_group(i).value_counts(subset="name")[0].item()):
        try:
            g1.iloc[j,1] = g2.iloc[j,1]
        except:
            g1.reset_index(drop=True).iloc[j,1] = np.nan
    g1.to_csv("./temp_file/{name}_target.csv".format(name=i), index=False)
```
